<p><font size="+1"> 2.Import Dense เพื่อกำหนดการ layer เพื่อให้ Neural เชื่อมต่อกันทุก Neural ยัง layer ถัดไป </font></p>

```python
#1.Import Sequential  เพื่อกำหนดลักษณะของชั้น Layer
#2.Import Dense เพื่อกำหนดการ layer เพื่อให้ Neural เชื่อมต่อกันทุก Neural ยัง layer ถัดไป
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

<p><font size="+1"> 1. Seq..ทำหน้าที่เหมือนสร้าง Layer ขึ้นมาโดยให้แต่หละ Layer มีการเชื่อมต่อกันหมด โดยมี neural ที่ 1 (Input) มี 1 neural และมีข้อมูล Input มา 1 จำนวน </font></p>
<p><font size="+1">2.model....เพื่อกำหนดว่า model มีการ Compile โดยการใช้ optimizer แบบ sgd : Gradient descent มีการคิดหรือคำนวณแบบ mean_squared_error </font></p>
<h2> <u>Gradient descent :</u></h2>
    <img src="Gradient descent.jpeg" width=300 height=187 />
<h2> <u>mean_squared_error equation : </u></h2>
    <img src="mean_squared_error equation.png" width=337 height=71/>



```python
#1. Seq..ทำหน้าที่เหมือนสร้าง Layer ขึ้นมาโดยให้แต่หละ Layer มีการเชื่อมต่อกันหมด โดยมี neural ที่ 1 (Input) มี 1 neural และมีข้อมูล Input มา 1 จำนวน 
model = Sequential([Dense(units=1,input_shape=[1])])
#2.model....เพื่อกำหนดว่า model มีการ Compile โดยการใช้ optimizer แบบ sgd : Gradient descent มีการคิดหรือคำนวณแบบ mean_squared_error
model.compile(optimizer='sgd',loss='mean_squared_error')
```

<p><font size="+1"> 1.สร้างข้อมูลแกน X ขึ้น เพื่อทำการ predict </font></p>
<p><font size="+1"> 2.สร้างข้อมูลแกน Y ขึ้น เพื่อทำการ predict </font></p>

```python
#1.สร้างข้อมูลแกน X ขึ้น เพื่อทำการ predict
xs=np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0],dtype=float)
#2.สร้างข้อมูลแกน Y ขึ้น เพื่อทำการ predict
ys=np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0],dtype=float)
```

<p><font size="+1"> 1.ทำการ Train โดยใช้ข้อมูลที่สร้างขึ้นจำนวน 500 รอบ </font></p>

```python
#1.ทำการ Train โดยใช้ข้อมูลที่สร้างขึ้นจำนวน 500 รอบ
model.fit(xs, ys, epochs=500)
```

<p><font size="+1"> 1.ทำการเริ่ม Predict โดยให้ค่า X เป็น 10 </font></p>

```python
#1.ทำการเริ่ม Predict โดยให้ค่า X เป็น 10
print('x=10,y=',model.predict([10.0]))
```

<p><font size="+1"> 1.ทำการ Train โดยใช้ข้อมูลที่สร้างขึ้นจำนวน 1500 รอบ </font></p>

```python
#1.ทำการ Train โดยใช้ข้อมูลที่สร้างขึ้นจำนวน 1500 รอบ
model.fit(xs, ys, epochs=1500)
```

<p><font size="+1"> 1.ทำการกำหนดค่า X เป็น 10 และให้ model ทำการ Predict </font></p>

```python
#1.ทำการกำหนดค่า X เป็น 10 และให้ model ทำการ Predict
print('x=10,y=',model.predict([10.0]))
```
